# **Matplotlib - PARTE 1**

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

# **Line Graph**

```python
x = np.array([1,2,3,4,5,6])
y = np.power(x,3)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,5))
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
plt.style.use("seaborn-v0_8-darkgrid")

# cambiar los colores de la lnea, ej. b==Blue (azul)
plt.figure(figsize=(10,5))
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color="b")
plt.show()
```

```python
# color rojo: r == red (rojo)
plt.figure(figsize=(10,5))
x = np.linspace(0,10,100)
y = np.sin(x)
# plt.plot(x,y,color="r")
plt.plot(x,y,"r-")          # Alternativa
plt.show()
```
